```python
# 3. Read the amount and percentage of interest from the keyboard and find final amount after adding interest in original amount.
print('Calculate the loan repayment amount using simple interest'))
amount = int(input('Enter the loan amount in INR:'))
interest_per = int(input('Enter the loan interest in %:'))
duration = int(input('Enter the loan duration in years:'))
loan = amount * (1+  interest_per * duration / 100)
print('Final loan repayment amount in INR:', loan)
```

```python
# 5. Write a Python program to calculate profit or loss. Input is selling cost and actual cost of a product.
sell = int(input('Enter the selling price of the product in INR:'))
buy = int(input('Enter the buying price of the product in INR:'))
pnl = sell - buy
if (pnl > 0):
  print("The profit on the product is: ", pnl)
elif (pnl<0):
  print("The loss on the product is: ", pnl)
else:
  print("There is neither profit  nor loss on the product.")
```

```python
#9. See the tuple given: num = (56,2,35,41,43,48,32,56,71,55,68)
#Write a program to create two lists from it where first list contains only even numbers and second will contain only odd numbers.
num = (56,2,35,41,43,48,32,56,71,55,68)
even = []
odd = []
for n in num:
  if n%2:
    odd.append(n)
  else:
    even.append(n)
print ('ODD numbers list:  ', odd)
print ('EVEN numbers list: ', even)
```

```python
# 12.Write a Python class named ‘Rectangle’ with two attributes length and width and
# a method named ‘calculate’ which will compute the area of a rectangle. Create an
# object to pass length and width and call the method to calculate the area of
# rectangle.

class Rectangle:
  def __init__(self, length, width):
    self.length = length
    self.width = width

  def area(self):
    return self.length*self.width;

len = int(input('Enter the length:'))
wid = int(input('Enter the width:'))
area = Rectangle(len,wid);
print ('Area of the rectangle is:', area.area())
```
