```python
path = '/content/drive/My Drive/ds-datasets/cricket.csv'
import pandas as pd
df = pd.read_csv(path)
df
```

```python
# 2. List the name of cricketer and their respective runs
df[['name', 'runs']]
```

```python
# 3. Find total wickets taken by them
totalWickets = df['wickets'].sum()
totalWickets
```

```python
# 4. Find average of catches taken
# Method 1
avgCatches = df['catches'].sum() / len(df.index)
print (avgCatches)
# Method 2
avgCatches = df['catches'].mean()
print (avgCatches)
```

```python
# 5. Find the name of wicketkeeper
keeper = df[df['stumpings'] > 0]
keeper['name'].values[0]
```

```python
# 6. print the name of bowler who played highest number of matches
maxMatches = df.query('matches == matches.max()' and 'wickets / matches > 1' )
maxMatches = maxMatches.sort_values(['matches'], ascending=False)
maxMatches['name'].values[0]
```

```python
# 7. Find average of all the bowlers
df_avgWicket = df
df_avgWicket['Avg_Wicket'] = df_avgWicket['wickets'] / df_avgWicket['matches']
df_avgWicket
```

```python
# 8. Find name of the bowler with least bowling average (wickets per match).
df_least_bowl_avg = df_avgWicket.query('Avg_Wicket > 0.0' )
df_least_bowl_avg = df_least_bowl_avg.sort_values(['Avg_Wicket'], ascending=True)
df_least_bowl_avg['name'].values[0]
```

```python
# 9. Print information about all players whose catches per match ratio is greater than 0.5 store it in a csv file.
df_catches_per_match = df[df['catches'] / df['matches'] > 0.5]
#df_catches_per_match
df_catches_per_match.to_csv('/content/drive/My Drive/ds-datasets/df_catches_per_match.csv')
```

```python
# 10. Draw the bar chart of playr name against their runs.
ax = df.plot.bar(x='name', y='runs', rot=0)
ax.tick_params(rotation=45)
```

```python
# 11. Sort and print information about players by ascending order of runs
df_sort_run = df.sort_values(by=['runs'], ascending=True)
df_sort_run
#sorted_df = df.sort_values(by=['Column_name'], ascending=True)
```

```python
# 12. print the names of players whose wickets are greater than matches
df_wickets_per_match = df[df['wickets'] / df['matches'] >= 1]
df_wickets_per_match
```
