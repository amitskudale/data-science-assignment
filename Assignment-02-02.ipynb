```python
# 1. Count how many number of posts are paid and unpaid?
paid_post = unpaid_post = null_post = 0
paid_post = df[df.Paid > 0].shape[0]
unpaid_post = df[df.Paid == 0].shape[0]
null_post = df[df.Paid.isnull()].shape[0]

print('Paid Posts: ', paid_post)
print('Unpaid Posts: ', unpaid_post)
print('Null Posts: ', null_post)
```

```python
# 3. Find average number of likes to the post in September month.
df_sept = df[df['Post Month']==9]
avgLikes = df_sept['like'].mean()
avgLikes
```

```python
# 4. Print the information about the posts which are links.
df_link = df[df['Type'] == 'Link']
df_link
```

```python
# 5. How many total shares are present for the video posts?
df_video = df[df['Type'] == 'Video']
video_shares = df_video['share'].sum()
video_shares
```

```python
# 6. In which hour of the day maximum posts are posted?
hourly_counts = df.groupby('Post Hour').size()
max_posts_hour = hourly_counts.idxmax()
max_posts_count = hourly_counts.max()
print('Post Hour:', max_posts_hour, 'Post count:', max_posts_count)
```

```python
# 7. In which day of October, maximum number of interactions are found?
df_october = df[df['Post Month'] == 10]
df_october
df_oct_days = pd.DataFrame()
df_oct_days = pd.DataFrame(columns=['Date', 'Count'])
df_oct_days
i = 1
while (i < 8):
  df_temp = df_october[df_october['Post Weekday'] == i]
  count = df_temp['Total Interactions'].sum()
  df2 = pd.DataFrame({'Day': [i], 'Count': [count]})  # Single-row DataFrame to add
  df_oct_days = df_oct_days._append(df2, ignore_index = True)
  i+=1
df_oct_days = df_oct_days.sort_values(['Count'], ascending=False)
#print(df_oct_days)
print(int(df_oct_days['Day'].values[0]), ':', int(df_oct_days['Count'].values[0]))
```

```python
# 8. Print the information of the posts where likes are above 200 and shares are above 30.
df_likes_shares = df.query('like > 200' and 'share  > 30' )
df_likes_shares
```

```python
# 9. Which kind of posts are posted maximum number of times in November month?
df_november = df[df['Post Month'] == 11]
df_nov_type = pd.DataFrame()
df_nov_type = pd.DataFrame(columns=['Type', 'Count'])
array = df_november['Type'].unique()
for n in array:
  df_temp = df_november[df_november['Type'] == n]
  count = df_temp.shape[0]
  df2 = pd.DataFrame({'Type': [n], 'Count': [count]})  # Single-row DataFrame to add
  df_nov_type = df_nov_type._append(df2, ignore_index = True)
df_nov_type = df_nov_type.sort_values(['Count'], ascending=False)
print (df_nov_type['Type'].values[0])
```

```python
# 10. Find average and median of interactions for the paid (1) posts.
df_paid = df[df['Paid'] == 1]
print ('Average:', df_paid['Total Interactions'].mean())
print ('Median: ', df_paid['Total Interactions'].median())
```
